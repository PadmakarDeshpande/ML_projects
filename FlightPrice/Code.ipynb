### Airfare Price Prediction

Dataset details: https://www.kaggle.com/code/vinayshaw/airfare-price-prediction/data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df_train = pd.read_excel('data/Data_Train.xlsx')
df_train.head()
```

```python
df_test = pd.read_excel('data/Test_set.xlsx')
df_test.head()
```

```python
# combining the data so that common pre-processing can be done
# later they will be separated 
df = df_train.append(df_test)
```

```python
df.info()
```

```python
df.tail()
```

```python
df['Date_of_Journey'].str.split('/')
```

```python
# Handle Date_of_Journey
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
df.info()
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
df.head()
```

```python
df['Arrival_Time'].str.split(' ').str[0]
```

```python
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
df
```

```python
df['Arrival_Hour'] = df['Arrival_Time'].apply(lambda x:x.split(':')[0])
df['Arrival_Mins'] = df['Arrival_Time'].apply(lambda x:x.split(':')[1])
```

```python
df['Arrival_Hour'] = df['Arrival_Hour'].astype(int)
df['Arrival_Mins'] = df['Arrival_Mins'].astype(int)
df.info()
```

```python
df.drop('Arrival_Time', axis=1, inplace=True)
```

```python
df['Dep_Hour'] = df['Dep_Time'].apply(lambda x:x.split(':')[0])
df['Dep_Mins'] = df['Dep_Time'].apply(lambda x:x.split(':')[1])
```

```python
df['Dep_Hour'] = df['Dep_Hour'].astype(int)
df['Dep_Mins'] = df['Dep_Mins'].astype(int)
df.info()
```

```python
df.drop('Dep_Time', axis=1, inplace=True)
```

```python
# Fix the Total_Stops and drop Route
df['Total_Stops'].unique()
```

```python
df[df['Total_Stops'].isnull()]
```

```python
# since NaN value has high duration, lets use 2 stops
df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0,
                                           '1 stop':1,
                                          '2 stops':2,
                                          '3 stops':3,
                                          '4 stops':4})
```

```python
df[df['Total_Stops'].isnull()].index
```

```python
df['Total_Stops'][9039] = 2
```

```python
df[df['Total_Stops'].isnull()]
```

```python
df.drop('Route', axis = 1, inplace=True)
df.info()
```

```python
# Convert Duration into mins
df['Duration'].unique()
```

```python
df['Duration_Hour'] = df['Duration'].str.split(' ').str[0].str.replace('h','')
```

```python
df[df['Duration_Hour']=='5m']
```

```python
# These values make no sense, better to drop these two entries entirly
df[df['Duration_Hour']=='5m'].index
```

```python
df.drop(df[df['Duration_Hour']=='5m'].index, axis=0, inplace=True)
```

```python
df[df['Duration_Hour']=='5m']['Duration_Hour'].sum()
```

```python
df['Duration_Hour'] = df['Duration_Hour'].astype(int)
```

```python
df.info()
```

```python
df['Duration_Mins'] = df['Duration'].str.split(' ').str[1].str.replace('m','')
df['Duration_Mins'].unique()
```

```python
df[df['Duration_Mins'].isnull()]
```

```python
# Looks like the Duration had only hours, hence fill the mins with 0
df['Duration_Mins'] = df['Duration_Mins'].fillna(0)
```

```python
df['Duration_Mins'] = df['Duration_Mins'].astype(int)
df['Duration_Mins'].unique()
```

```python
df['Duration_Mins'] += df['Duration_Hour']*60
df['Duration_Mins'].unique()
```

```python
df.drop(['Duration','Duration_Hour'], axis=1, inplace=True)
```

```python
df.info()
```

```python
df['Additional_Info'].unique()
```

```python
df.drop('Additional_Info',axis=1, inplace=True)
```

```python
df['Airline'].unique()
```

```python
df['Source'].unique()
```

```python
df['Destination'].unique()
```

```python
# Lets encode these columns
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['Airline'] = le.fit_transform(df['Airline'])
df['Source'] = le.fit_transform(df['Source'])
df['Destination'] = le.fit_transform(df['Destination'])
```
